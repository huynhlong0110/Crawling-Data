##**1. Collecting data**

Collecting the daily stock prices (closing prices) of 60 stock symbols from Ho Chi Minh Stock Exchange (HOSE) in 2021.

daterange function is used to initialize time

```python
def daterange(start_date, end_date): 
    for n in range(int((end_date - start_date).days)):
        yield start_date + timedelta(n)
```

list_of_date function is used to format Date by dd/mm/yyyy

```python
def list_of_date (start_date,end_date):
  List_Date = []
  for single_date in daterange(start_date, end_date):
      List_Date.append(single_date.strftime("%d/%m/%Y"))    
  return List_Date
```

Create converNumber function to convert data from string type to float type

```python
def convertNumber(number): 
  a = number.replace(",", ".")
  b = float(a)
  return b
```

Create two list to store data from Stockbiz website

```python
Close_Price= [] 
listDate = [] 
```

Write a function crawl_daily_stock_prices that uses the Beautiful Soup library to collect data on the closing price of a stock ticker by day on the StockBiz page by following the link:
'https://www.stockbiz.vn/Stocks/'+name+'/LookupQuote.aspx?Date='+date

**name**: Is the stock code name.

**date**: Is the date.

Then, process the data and return the results in the form of a tupple including: **daily data** and **closing price data** of that day.

```python

def crawl_daily_stock_prices(name,date):  
    URL = 'https://www.stockbiz.vn/Stocks/'+name+'/LookupQuote.aspx?Date='+date
    soup = BeautifulSoup(requests.get(URL).text, 'html.parser')
    dateTime = soup.find('div', {'class':'CompanyTitle'})
    closePrice = soup.find('div', attrs={'class':'Quote'})
    listPrice = [] 
    dateItem = dateTime.get_text().split()
    closePrice = closePrice.get_text().split()
    for u in dateItem:
      if not u.isalpha():  
        listDate.append(u)       
    
    for u in closePrice:
      if not u.isalpha():  
        listPrice.append(u)
    price = listPrice[0]

    prices = convertNumber(price)

    Close_Price.append(prices)
    return (listDate, Close_Price)
    
```
